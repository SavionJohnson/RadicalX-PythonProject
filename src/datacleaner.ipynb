#Grouping/Eliminating Columns

```python
# creates simplified_df, which will be used for grouping users
simplified_df = df[["name","timezone","region","language","priorexperience"]]
simplified_df
```

```python
for i, j in (simplified_df.language.items()):
    keywords = str(j).lower().split(" ")
    words = []
    if "english" in keywords:
        words.append("English")
    if "spanish" in keywords:
        words.append("Spanish")
    if "french" in keywords:
        words.append("French")
    if "chinese" in keywords:
        words.append("Chinese")
    if "hindi" in keywords:
        words.append("Hindi")
    if "bengali" in keywords:
        words.append("Bengali")
    if "portuguese" in keywords:
        words.append("Portuguese")
    if "japanese" in keywords:
        words.append("Japanese")
    if "telegu" in keywords:
        words.append("Telegu")    
    if "italian" in keywords:
        words.append("Italian")
    if "vietnamese" in keywords:
        words.append("Vietnamese")
    if "korean" in keywords:
        words.append("Korean")
    if "tamil" in keywords:
        words.append("Tamil")
    if "gujurati" in keywords:
        words.append("Gujurati")    
    if "urdu" in keywords:
        words.append("Urdu")
    if "turkish" in keywords:
        words.append("Turkish")   
    
    simplified_df.at[i,'language'] = ",".join(words)
simplified_df
```

```python
# clean simplified_df by removing any nan values and splitting rows 
cleaned_df1 = simplified_df.assign(language=simplified_df.language.str.split(","), timezone=simplified_df.timezone.str.split(","), region=simplified_df.region.str.split(",")).explode(["language"])
cleaned_df2 = cleaned_df1.explode('timezone')
cleaned_df3 = cleaned_df2.explode('region')
cleaned_df3.timezone.fillna("Unknown", inplace = True)
cleaned_df3.region.fillna("Unknown", inplace = True)
cleaned_df3.language.fillna("Unknown", inplace = True)
cleaned_df3.priorexperience.fillna("Unknown", inplace = True)
cleaned_df3
```

```python
grouped_df = simplified_df.groupby(["timezone", "region", "language", "priorexperience"])
for key, item in grouped_df:
    print(grouped_df.get_group(key), "\n\n")
    
grouped_df.get_group(("Americas", "Americas", "English", "Yes"))
```
