```python
# creates simplified_df, which will be used for grouping users
simplified_df = df[["name","timezone","region","language","priorexperience"]]
simplified_df
```

```python
# clean simplified_df by removing any nan values and splitting rows 
simplified_df = simplified_df.assign(timezone=df.timezone.str.split(",")).explode('timezone')
simplified_df = simplified_df.assign(region=df.region.str.split(",")).explode('region')
simplified_df.timezone.fillna("Unknown")
simplified_df.region.fillna("Unknown")
simplified_df.language.fillna("Unknown")
simplified_df.priorexperience.fillna("No")
simplified_df = simplified_df.replace("Any/No preference","Unknown")
```
